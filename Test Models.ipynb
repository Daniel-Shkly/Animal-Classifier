```python
def Three_Class_Pred(model, img_path):
  img = load_img(img_path)
  img = img_to_array(img)[:,:,:]
  img = np.expand_dims(img ,axis=0)
  img = tf.image.resize(img, (128, 128))

  img = img * (1./255)

  x = model.predict(img).argmax()
  #y = model.predict(img)

  if x == 0:
    return 0
  elif x == 1:
    return 1
  elif x == 2:
    return 2
```

```python
def Four_Class_Pred(model, img_path):
  img = load_img(img_path)
  img = img_to_array(img)[:,:,:]
  img = np.expand_dims(img ,axis=0)
  img = tf.image.resize(img, (128, 128))

  img = img * (1./255)

  x = model.predict(img).argmax()
  #y = model.predict(img)

  if x == 0:
    return 0
  elif x == 1:
    return 1
  elif x == 2:
    return 2
  elif x == 3:
    return 3
```

### 3 Class Test

```python
# The model to be tested
model = CDNmodel

# Test Directories
c_zero_dir = CatDir
c_one_dir = DogDir
c_two_dir = NeitherDir

# Failure and Success Counts - to get examples
s = 0
f = 0
c0_s = []
c0_f = []

c1_s = []
c1_f = []

c2_s = []
c2_f = []

# Class 0 Test
c_zero = 0
c_zero_p = 0
for filename in os.listdir(CatDir):
  file = os.path.join(c_zero_dir, filename)

  if Three_Class_Pred(model, file) == 0:
    c_zero += 1

    if s < 2:
      c0_s.append(file)
      s += 1

  elif f < 2:
    c0_f.append(file)
    f += 1

  c_zero_p += 1

acc_c_zero = c_zero/c_zero_p

s = 0
f = 0
# Class 1 Test
c_two = 0
c_two_p = 0
for filename in os.listdir(DogDir):
  file = os.path.join(c_one_dir, filename)

  if Three_Class_Pred(model, file) == 1:
    c_two += 1

    if s < 2:
      c1_s.append(file)
      s += 1

  elif f < 2:
    c1_f.append(file)
    f += 1

  c_two_p += 1

acc_c_two = c_two/c_two_p

s = 0
f = 0
# Class 2 Test
c_three = 0
c_three_p = 0
for filename in os.listdir(NeitherDir):
  file = os.path.join(c_two_dir, filename)
  
  if Three_Class_Pred(model, file) == 2:
    c_three += 1

    if s < 2:
      c2_s.append(file)
      s += 1

  elif f < 2:
    c2_f.append(file)
    f += 1

  c_three_p += 1

acc_c_three = c_three/c_three_p

print('Class 0 accuracy: ', acc_c_zero)
print('Class 1 accuracy: ', acc_c_two)
print('Class 2 accuracy: ', acc_c_three)

print('\nClass 0\nS: ', c0_s, '\nF: ', c0_f, '\n')
print('Class 1\nS: ', c1_s, '\nF: ', c1_f, '\n')
print('Class 2\nS: ', c2_s, '\nF: ', c2_f, '\n')
```
